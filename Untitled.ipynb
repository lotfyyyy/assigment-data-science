```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.figure(figsize= (25,20))
plt.subplot(1,1,1)
sns.boxplot(x= netflix_df.release_year, color='red')
plt.show()            
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.figure(figsize= (30,25))
plt.subplot(1,1,1)
plt.hist(netflix_df.release_year, color='green', edgecolor = 'black', alpha = 1)
plt.xlabel('release_year')
plt.show()
```

```python
import scipy.stats as stats
Skewness = pd.DataFrame({'Skewness' : [stats.skew(netflix_df.release_year)]},
                        index=['release_year'])
Skewness
```

```python
import seaborn as sns 
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="type",data=netflix_df)
```

```python
from sklearn.preprocessing import LabelEncoder
import copy
netflix_df_encoded = copy.deepcopy(netflix_df)
netflix_df_encoded.loc[:,['release_year']] = netflix_df_encoded.loc[:,['release_year']].apply(LabelEncoder().fit_transform) 
sns.pairplot(netflix_df_encoded)  
plt.show()
```
